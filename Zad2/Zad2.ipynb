```
import sklearn;
import numpy as np;
import pandas as pd;
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set()

data = {
'size':['S', 'M', 'L', 'XL', 'XXL', 'M',np.nan, 'M', 'XXL', 'XL', 'S', np.nan],
'color':['green', 'red', 'blue', 'white',np.nan, 'dark', 'yellow', 'grey', 'white', 'dark', 'yellow', np.nan],
'gender':['female','male', 'male', 'male',np.nan, 'female', 'female', 'male', 'female', 'male', 'female', np.nan],
'price':[110.0, 120.0, np.nan, 140.0,np.nan, 150.0, 160.0, 170.0, 180.0, 190.0, 200.0, np.nan],
'weight':[250, 275, 300, 325, 350,np.nan, 375, 400, 425, 450, 475, np.nan],
'bought':['yes', 'no', 'no','yes',np.nan,'yes','yes','no','no','yes','no', np.nan]
}

df_raw = pd.DataFrame(data=data)
df = df_raw.copy()
df
```

```
df.dropna(thresh=len(df.keys())-2)
```

```
df['price'].fillna(df['price'].mean(),inplace=True)
df
```

```
df['weight'].fillna(df['weight'].interpolate(),inplace=True)
df
```

```
df['size'].fillna('unisex',inplace=True)
df['color'].fillna('black',inplace=True)
df['gender'].fillna('male',inplace=True)
df['bought'].fillna('yes', inplace=True)
df
```

```
df['cost'] = pd.cut(df['price'], bins=3, labels=('cheap', 'normal_price', 'expensive'))
df
```

```
df['bought_yes'] = df['bought'].apply(lambda x: 1 if 'yes' in x else 0)
df
```

```
from sklearn.model_selection import train_test_split

df_in=df[['price', 'gender', 'color']]
df_out=df[['bought_yes']]

x_train, x_test, y_train, y_test = train_test_split(df_in, df_out)

x_train, y_train
```
