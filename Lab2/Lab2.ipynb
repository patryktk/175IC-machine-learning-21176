```
import pandas as pd


pd.read_csv("samochody1tys.csv");
data = pd.read_csv("samochody1tys.csv");

#Wyświetlenie pierszych 5 rekordów
print(data.head(5))

#Wyświetlenie ostatnich 5 rekordów
print(data.tail(5))

#Wyświetlenie 5 przypadkowych rekordów
print(data.sample(n=5))

#Wyświetlenie najdroższego samochodu
print(data.sort_values('cena', ascending=False).head(1))

#Usuwanie kolumn które mają pusty rekord (N/A)
print(data.dropna())

#Wszystkie puste rekordy (N/A) zapełnia wiadomościa 'puste'
print(data.fillna("puste"))

#Wszystkie samochody z silnikiem Diesel
silnik = data[data.rodzaj_silnika=="Diesel"]
print(silnik)

#Wszystkie samochody wyprodukowane po 2012 roku
rokprodukcji = data[data.rok_produkcji>2012]
print(rokprodukcji)

#Usunięcie kolumny id
print(data.drop(columns= 'id'))

#samochód z najmniejszym przebiegiem
print(data.nsmallest(1,'przebieg'))
```
